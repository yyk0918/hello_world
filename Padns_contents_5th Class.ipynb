```python
import numpy as np
import pandas as pd
from pandas import DataFrame, Series
```

```python
movies = pd.read_csv('http://bit.ly/imdbratings')
movies
```

```python
# 행과 열의 수 확인
movies.shape
```

```python
# 먼저, 영화상영시간이 200분 이상이면 true, 200분이하면 false라고 표시해주는 리스트를 생성해야함
booleans = []
for length in movies.duration:
    if length >= 200:
        booleans.append(True)
    else:
        booleans.append(False)
```

```python
booleans
```

```python
# 리스트에 있는 값의 수가 전체 영화수가 일치하는지 확인
len(booleans)
```

```python
# 1-5번영화먼저 확인
booleans[0:5]
```

```python
# 데이터프레임을 만들기전에 리스트를 Series로 변환
is_long = pd.Series(booleans)
is_long.head()
```

```python
# use bracket notation with the boolean Series to tell the DataFrame which rows to display
movies[is_long]
```

```python
# 위과정을 아래와 이 간편하게도 가능함
is_long = movies.duration >= 200
movies[is_long]

# is_long을 생성하지 않고 더 간단하게!
movies[movies.duration >= 200]
```

```python
# Seriese에서 장르만 선택하기 
movies[movies.duration >= 200].genre
```

```python
# loc내장함수를 이용하는 방법
movies.loc[movies.duration >= 200, 'genre']
```

```python
# CORRECT: use the '&' operator to specify that both conditions are required
movies[(movies.duration >=200) & (movies.genre == 'Drama')]
```

```python
# use the '|' operator to specify that a row can match any of the three criteria
movies[(movies.genre == 'Crime') | (movies.genre == 'Drama') | (movies.genre == 'Action')].head(10)
```

```python
# 각나라별 음주현황데이터 불러오기
drinks = pd.read_csv('https://raw.githubusercontent.com/justmarkham/pandas-videos/master/data/drinks.csv')
drinks.head()
```

```python
drinks
```

```python
# 전체 데이터의 맥주 평균 소비량
drinks.beer_servings.mean()
```

```python
# 아프리카 대륙의 평균 소비량
drinks[drinks.continent=='Asia'].wine_servings.mean()
```

```python
# 각 대륙별 평균 소비량
drinks.groupby('continent').beer_servings.mean()
```

```python
# 다양한 통계치를 한꺼번에!
drinks.groupby('continent').beer_servings.agg(['count', 'mean', 'min', 'max'])
```

```python
# 쥬피터 노트북에 그래프를 보이게 하려면 필수!
%matplotlib inline
```

```python
# 그래프를 그려보자!
drinks.groupby('continent').mean().plot(kind='bar')
```
